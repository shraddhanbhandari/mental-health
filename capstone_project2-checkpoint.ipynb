```python
from nltk.corpus import stopwords
import re
import emoji
import string
```

```python
def remove_html(text):
    pattern = re.compile('\\')
    return pattern.sub(r'', text)
```

```python
def remove_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251" 
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
abcd = stopwords.words('english')
words_to_remove = ['not','wouldn','mustn',"wouldn't","don't","didn't","aren't",'couldn','aren',"doesn't","hasn't","won't","haven't","hadn't","mightn't","wasn't","isn't","needn't",'mightn',"isn","shouldn't",'ain',"weren't","couldn't","mustn't"]
abcd = [word for word in abcd if word not in words_to_remove]
abcd
```

```python
def text_process(text):
    """
    1. remove the punctuation
    2. remove the stopwords
    3. return the clean text words
    """
    nopunc = [char for char in text if char not in string.punctuation]
    nopunc = "".join(nopunc)
    return[word for word in nopunc.split() if word not in abcd]
    
```

# LSTMS

```python
import gensim
import nltk
import re
import pandas as pd
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding
```

```python
mh = pd.read_csv(r"Suicide_Detection.csv")
```

```python
mh = mh[['text','class']]
```

```python
mh['class'].replace({'suicide':1, 'non-suicide':0}, inplace=True)
```

```python
mh_x =mh.iloc[:, 0]
mh_y =mh.iloc[:, 1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(mh_x, mh_y, test_size=.2)
```

```python
y_train = to_categorical(y_train)
```

```python
max_num_words = 10000      # max words to be taken from entire corpus
seq_len = 50               # 50 words each from each document
embedding_size = 100       
```

```python
from keras.preprocessing.text import Tokenizer         # is used for tokenization
from keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer(num_words=max_num_words)
tokenizer.fit_on_texts(mh.text)
x_train = tokenizer.texts_to_sequences(x_train)  # will convert the text into sequence ids
x_train = pad_sequences(x_train, maxlen= seq_len)

x_test = tokenizer.texts_to_sequences(x_test)
x_test = pad_sequences(x_test, maxlen=seq_len)

model = Sequential()
model.add(Embedding(input_dim = max_num_words,
                    input_length=seq_len,
                    output_dim = embedding_size))
```

```python
model.add(LSTM(5))
model.add(Dense(2, activation='softmax')) # 2 becoz 2 class in the target variablee
from tensorflow.keras.optimizers import Adam
adam = Adam(lr = .001)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=5)
```

```python
pred_mh = model.predict(x_test)
pred_mh = pred_mh.argmax(axis = 1)
```

```python
tab_lstm = confusion_matrix(y_test, pred_mh)
tab_lstm
```

```python
accuracy_score(y_test, pred_mh)
```

```python
with open('lstm_mh', 'wb') as f:
    pickle.dump(model, f)
```

```python
with open('lstm_mh', 'rb') as f:
    lstm = pickle.load(f) 
```

# Bi-directional lstms

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(mh_x, mh_y, test_size=.2)
```

```python
y_train = to_categorical(y_train)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dense
```

```python
tokenizer = Tokenizer(num_words=max_num_words)
tokenizer.fit_on_texts(mh.text)
x_train = tokenizer.texts_to_sequences(x_train)  # will convert the text into sequence ids
x_train = pad_sequences(x_train, maxlen= seq_len)

x_test = tokenizer.texts_to_sequences(x_test)
x_test = pad_sequences(x_test, maxlen=seq_len)

model = Sequential()
model.add(Embedding(input_dim = max_num_words,
                    input_length=seq_len,
                    output_dim = embedding_size))
model.add(Bidirectional(LSTM(64)))
```

```python
model.add(Dense(2, activation='softmax')) # 2 becoz 2 class in the target variablee
from tensorflow.keras.optimizers import Adam
adam = Adam(lr = .001)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=1)
```

```python
pred_mhbi = model.predict(x_test)
pred_mhbi = pred_mhbi.argmax(axis = 1)
```

```python
tab_bilstms = confusion_matrix(y_test, pred_mhbi)
tab_bilstms
```

```python
accuracy_score(y_test, pred_mhbi)
```

```python
with open('bilstm_mh', 'wb') as f:
    pickle.dump(model, f)
```
